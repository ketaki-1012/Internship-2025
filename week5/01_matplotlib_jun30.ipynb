```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([12, 23, 32, 37, 44, 49, 51, 56, 59, 64])
y = np.array([10, 14, 19, 25, 30, 35, 40, 43, 51, 59])
z = np.array([70, 65, 61, 58, 50, 48, 42, 37, 28, 15])
```

```python
#Line plot
```

```python
plt.title("Line plot")
plt.xlabel("x values")
plt.ylabel("Y values")
plt.plot(x, y, color = 'red', marker = 'H', mfc = 'blue', ls = '--');
plt.grid()
```

```python
plt.title("Line plot")
plt.xlabel("x values")
plt.ylabel("z values")
plt.plot(x, z, color = 'black', marker = 'o', mfc = 'red', ls = '--');
plt.grid()
```

```python
plt.title("Line plot")
plt.xlabel("x values")
plt.ylabel("Y values")

plt.plot(x, y, color = 'red', marker = 'H', mfc = 'blue', ls = '--',label = '2023');
plt.plot(x, z, color = 'black', marker = 'o', mfc = 'red', ls = '--',label = '2024');

plt.grid()
plt.legend();
```

```python
plt.figure(figsize=(16,9))
plt.subplot(2, 2, 1)
plt.title("Line plot")
plt.xlabel("x values")
plt.ylabel("Y values")
plt.plot(x, y, color = 'red', marker = 'H', mfc = 'blue', ls = '--');
plt.grid()

plt.subplot(2, 2, 2)
plt.title("Line plot")
plt.xlabel("x values")
plt.ylabel("z values")
plt.plot(x, z, color = 'black', marker = 'o', mfc = 'red', ls = '--');
plt.grid()
```

```python
df = pd.read_csv("Social_Network_Ads.csv")  # Use your actual file name

# Count how many purchased (1) and not purchased (0)
counts = df['Purchased'].value_counts().sort_index()
labels = ['Not Purchased', 'Purchased']

plt.figure(figsize=(10, 8))

# 1. Bar Chart
plt.subplot(2, 2, 1)
plt.bar(labels, counts, color=['red', 'green'])
plt.title("Bar Chart")

# 2. Pie Chart
plt.subplot(2, 2, 2)
plt.pie(counts, labels=labels, autopct='%1.0f%%', colors=['red', 'green'])
plt.title("Pie Chart")

# 3. Histogram of Salary
plt.subplot(2, 2, 3)
plt.hist(df['EstimatedSalary'], color='blue')
plt.title("Salary Histogram")

# 4. Scatter Plot: Age vs Salary
plt.subplot(2, 2, 4)
plt.scatter(df['Age'], df['EstimatedSalary'], color='purple')
plt.title("Age vs Salary")

plt.tight_layout()
plt.show()
```
